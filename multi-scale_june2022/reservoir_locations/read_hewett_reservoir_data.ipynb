```python
import math
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
g=9.806         #standard gravity
cc=0.622        #contraction coefficient
cq=0.385        #weir coefficient

Hb=0.25         #height of the bottom of the orifice
diam=0.50       #diameter of the orifice
R=diam/2
Ac=math.pi*(diam/2)**2
Hspill=2.00     #height of the weir crest
Hdam=2.50       #height of the dam (unused)
L=10            #length (or width) of the weir crest
step=0.25
```

```python
def defineDepth(damIDList, maxHList, step):
    depthDict={}
    for damID in damIDList:
        depthDict[damID]=[]
        i=0
        while i<=5:
            depthDict[damID].append(i)
            i+=step
    return depthDict


def calcQ(h):
    ###DISCLAIMER###
    ###This part is based on formulas to calculate the discharge from the dam
    ###Formulas have to be modified according to the design used for the dam
    ###and to the most reliable sources and methods (e.g. USACE)
    if (h<=Hb):
        Q=0
    elif (h<(Hb+diam)):
        theta=2*math.acos(1-(h-Hb)/R)
        Ap=0.5*(R**2)*(theta-math.sin(theta))        
        Q=Ap*math.sqrt(2*g*(h-Hb)) #cc
        print(h,theta,Ap,Q)
    elif(h<Hspill):
        Q=cc*Ac*math.sqrt(2*g*h)
    else:
        Q=(cc*Ac*math.sqrt(2*g*h))+(cq*L*(math.sqrt(2*g))*(h-Hspill)**1.5)
    return Q
```
